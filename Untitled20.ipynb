```python
sns.histplot(df['MonthlyCharges'], kde=True)
plt.show()
```

### Monthly Charges Analysis

- MonthlyCharges ranges from approximately 18 to 120.
- The distribution is bimodal, indicating two major customer groups.
- A large number of customers pay around 20–30 per month.
- Another significant group pays around 70–100 per month.
- The feature is not normally distributed.

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()
```

```python
sns.histplot(df['tenure'], kde=True)
plt.show()
```

### Tenure Analysis

- Tenure ranges from 0 to 72 months.
- The distribution is bimodal.
- A large number of customers are either very new or very loyal.
- Fewer customers fall in the mid-tenure range.
- Tenure is expected to be an important factor influencing customer churn.

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.show()
```

```python
sns.histplot(df['SeniorCitizen'], kde=True)
plt.show()
```

### Senior Citizen Analysis

- The majority of customers are not senior citizens.
- Senior citizens represent a smaller portion of the customer base.
- Since SeniorCitizen is a binary categorical feature, count plots are more appropriate than histograms.
- Further analysis is needed to determine its relationship with churn.

```python
sns.boxplot(x='Churn', y='SeniorCitizen', data=df)
plt.show()
```
